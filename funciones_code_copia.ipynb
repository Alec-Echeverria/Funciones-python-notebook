# Funciones algebraicas


<h2 style="color:#8BC34A;"><b>Funciones algebraicas lineales</b></h2>

Tiene la forma de $$f(x)=mx + b$$ donde $m$ y $b$  $\in R$. 

$m$ puede ser calculada por: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

y $b$ es el punto de corte con el eje $y$. Su dominio es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = (-\infty, \infty)$



```python
#n es la resolución o puntos de la grafica
n= 50
m=10 #m es la pendiente
b=5  #b es donde corta la grafica

#definimos la funcion algebraica
def f(x):
    return m*x+b

#Creamos el dominio de la función => x es una varibale discreta
x=np.linspace(-10,10,num=n)

#Función
y=f(x)

#graficamos
fig, ax=plt.subplots()
ax.plot(x,y)
ax.grid()
plt.show()
```


<h2 style="color:#8BC34A;"><b>Funciones algebraicas polinómicas</b></h2>

Tiene la forma de $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{0}$$

a una función que tiene esta forma se le llama polinomio de grado $n$. A los elementos $a$ los llamaremos coeficientes donde $a \in R$. 

**Por ejemplo:**

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$

que es un polinomio de grado 7.

```python
#resolución
n= 50


#Funcion polinomica
def f(x):
    return (2*x**7)-(x**4)+(3*x**2)+4


#Creamos el dominio de la función => x es una varibale discreta
x=np.linspace(-10,10,num=n)


#definimos variable dependiente
y=f(x)

#graficamos
plt.plot(x,y)
plt.grid()
plt.show()
```

# Funciones trascendentes

<h2 style="color:#8BC34A;"><b>Funciones trigonométricas</b></h2>

Algunos ejemplos son las funciones $cos(x)$, $sen(x)$ y $tan(x)$


```python
#resolución
n= 100
#Creamos el dominio de la función => x es una varibale discreta
x=np.linspace(-10,10,num=n)

#Creamos funcion trigonometrica

def f(x):
    return (np.sin(x))

#definimos variable dependiente
y=f(x)

#graficamos
plt.plot(x,y)
plt.grid()
plt.show()
```

<h2 style="color:#8BC34A;"><b>Función exponencial</b></h2>

Tienen la forma de $$f(x)=a^x$$ donde la base $a$ es una constante positiva. Un gran ejemplo de una función exponencial es usando la base como el número de euler:

$$f(x)=e^x$$

```python
#resolución
n= 100
#Creamos el dominio de la función => x es una varibale discreta
x=np.linspace(-10,10,num=n)

#Creamos funcion exponencial

def f(x):
    return (np.exp(x))

#definimos variable dependiente
y=f(x)

#graficamos
plt.plot(x,y)
plt.grid()
plt.show()
```

<h2 style="color:#8BC34A;"><b>Función logaritmo</b></h2>


El logaritmo está definido por la **relación**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$ 

donde: 



*   $b$ es la base.
*   $n$ es el exponente al que está elevado la base.
*   $x$ es el resultado de elevar la base $b$ al exponente $n$

**Ejemplo:**

Teniendo b=2 y n=8, entonces:

$$2^8=256$$

Por lo que $x=256$. Calculando el logaritmo base 2 de $x$ es:

$$log_{2}(256) = 8$$


```python
#resolución
n= 100
#Creamos el dominio de la función => x es una varibale discreta => x debe ser mayor a 0
x=np.linspace(0.1,256,num=n)

#Creamos funcion logaritmica

def f(x):
    return (np.log2(x))

#definimos variable dependiente
y=f(x)

#graficamos
plt.plot(x,y)
plt.grid()
plt.show()

#en la grafica se detalla el resultado 8
```

# Funciones seccionadas

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
#resolución
n= 100
#Creamos el dominio de la función => x es una varibale discreta => x debe ser mayor a 0
x=np.linspace(-10,10,num=n)

#Creamos funcion seccionada

def h(x):
    #genamos una Y temporal y alamezamos los ceros alli
    Y=np.zeros(len(x))
    print("Y temporal en ceros:\n",Y)
    for indx,x in enumerate(x):
        if x>=0:
            Y[indx]=1
    print("valores convertidos a 1 en el vector Y temporal: \n",Y)
    return Y
    
    

#definimos variable dependiente
y=h(x)

#graficamos
plt.plot(x,y)
plt.grid()
plt.show()

#en la grafica se detalla el resultado 8
```

# Funciones compuestas


Una función compuesta se obtiene cuando se pasa como argumento de la función principal el valor de otra función, en este ejemplo tenemos:

$$f(x)=x^2$$
$$g(x)=x+1$$

se busca el valor de $g(f(x))$ siendo $x=3$

![](https://static.platzi.com/media/user_upload/function_composite_fx-72da81a5-31c8-460e-9ff9-efdd19c00400.jpg)

```python
#definimos la resolucion de los datos de x
n=500

#Creamos el dominio de la función => x es una varibale discreta => x debe ser mayor a 0
x=np.linspace(-10,10,n)

#definimos f(x)

def f(x):
    return x**2

#definimos g(x)
def g(x):
    return np.sin(x)


g_o_f = g(f(x))


plt.plot(x,g_o_f)
plt.show()
```

# Dependencias


<h2 style="color:#8BC34A;"><b>Desplazamientos verticales y horizontales</b></h2>


Siendo $c$ una constante mayor que cero, entonces la gráfica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
#definimos resolución
n= 100

#Creamos el dominio de la función => x es una varibale discreta => x 
x=np.linspace(-100,100,n)

#Definimos constante de desplazamiento
c=10

#definimos f(x)
def f(x):
    return x**2


#Definimos la variable dependiente y llamamos la constante de desplazamiento
y = f(x)


fig, ax = plt.subplots()

plt.plot(x,y)
plt.grid()

#Pintamos los ejes Axiales
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```


<h2 style="color:#8BC34A;"><b>Alargamientos y compresiones</b></h2>


Siendo $c$ una constante mayor que cero, entonces la gráfica:

- $y=c \cdot f(x)$ alarga la gráfica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la gráfica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la gráfica horizontelmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la gráfica horizontelmente en un factor de $c$.

```python
#definimos resolución
n= 1000

#Creamos el dominio de la función => x es una varibale discreta => x 
x=np.linspace(-20,20,n)

#Definimos constante de alargamamiento/ comprensión
c=2

#definimos f(x)
def f(x):
    return np.sin(x)


#Definimos la variable dependiente y llamamos la constante de desplazamiento
y = f(x)


fig, ax = plt.subplots()

plt.plot(x,y)
plt.grid()

#Pintamos los ejes Axiales
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

<h2 style="color:#8BC34A;"><b>Reflexiones</b></h2>


- $y=-f(x)$ refleja la gráfica respecto al eje x.
- $y=f(-x)$ refleja la gráfica respecto al eje y.

```python
#definimos resolución
n= 1000

#Creamos el dominio de la función => x es una varibale discreta => x 
x=np.linspace(-10,10,n)


#definimos f(x)
def f(x):
    return (x)**3


#Definimos la variable dependiente 
y = f(x)


fig, ax = plt.subplots()

plt.plot(x,y)
plt.grid()

#Pintamos los ejes Axiales
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
plt.show()
```

<h2 style="color:#8BC34A;"><b>Funciones de activación</b></h2>

Sirven para darles diferentes valores de salida a todo el proceso que surge de la combinación lineal (Unión sumadora) en el perceptrón.

En el ámbito de la inteligencia artificial, se conoce como función de activación a aquella función matemática que activa los pesos de una neurona en función de su resultado. Esto permite incorporar no linealidad en las redes neuronales. Existen diversas clases de funciones de activación que se emplean en diferentes aplicaciones de la inteligencia artificial, no limitándose únicamente al campo del Deep Learning.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-5,5, num=N)
```

## Función lineal
$$y=mx+b$$

```python
#Función lineal: mantener valores a lo largo de un proceso, predecir venta

def f(x):
  return x

plt.plot(x, f(x))
plt.grid()
```

## Función escalón o de Heaviside
$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
#Función escalón o de Heaviside: clasificaciones categóricas, ej. si está apagado o prendido, si existe o no existe. Utiliza valores binarios.

def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    

N=1000


y = H(x)

plt.plot(x,y)
plt.grid()
```

## Función sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

```python
#Función sigmoide: regresión logística, sirve cuando hablamos de probabilidades. Su rango va de 0 a 1. Cuando un valor es 0, su función será 0.5
def f(x):
  return 1/(1 + np.exp(-x))
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Función tangente hiperbólica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

```python
#Función tangente hiperbólica: conocida como función de escalamiento, va de -1 a 1.
def f(x):
  return np.tanh(x)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Función ReLU

$$R(x)=max(0,x)$$

```python
#Función ReLU: simulación de “neuronas muertas”

def f(x):
  return np.maximum(x,0)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```
